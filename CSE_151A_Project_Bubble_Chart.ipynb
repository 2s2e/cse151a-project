```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('spotify-and-youtube/Spotify_Youtube.csv')
```

```python
numerical_columns = df.select_dtypes(include=['number']).columns.tolist()
categorical_columns = df.select_dtypes(include=['object', 'category']).columns.tolist()

print(f"Numerical Columns: {len(numerical_columns)}")
print(numerical_columns)
print(f"Categorical Columns: {len(categorical_columns)}")
print(categorical_columns)
print(f"Total Columns: {len(df.columns)}")
```

```python
df_only_numerical = df[numerical_columns]
df_only_categorical = df[categorical_columns]
```

```python
import plotly.express as px

dff = df.groupby('Artist').agg({'Views': 'sum', 'Likes': 'sum'}).reset_index()

print(dff)
```

```python

fig = px.scatter(dff, x="Views", y="Likes",
	         size=df['Artist'].value_counts(),
                 hover_name="Artist", log_x=True, size_max=60)
fig.show()
```
