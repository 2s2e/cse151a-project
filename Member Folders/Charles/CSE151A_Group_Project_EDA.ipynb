```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew

df = pd.read_csv('Spotify_Youtube.csv')

numerical_features = ["Danceability", "Energy", "Loudness", "Speechiness", "Acousticness", "Instrumentalness",
              "Liveness", "Valence", "Tempo", "Duration_ms", "Views", "Likes", "Comments", "Stream"]

numerical_data = df[numerical_features]
```

```python
# Sub-samples for observation purpose
numerical_data.sample(n=10)
```

```python
def generate_histogram(column):
  sns.histplot(data=numerical_data, x=column)
  plt.title(column)
  plt.xlabel('Value')
  plt.ylabel('Frequency')
  plt.show()

for column in numerical_features:
  generate_histogram(column)
```

```python
from scipy.stats import skew

def is_heavily_skewed(column):
    """
    Determine if a column is heavily skewed.
    """

    skewness_value = skew(column, bias=False)
    is_skewed = abs(skewness_value) > 2

    return is_skewed, skewness_value

for column in numerical_features:
  is_skewed, skewness_value = is_heavily_skewed(df[column])
  if is_skewed:
    print(f"{column} is heavily skewed, skewness value: {skewness_value}")
  else:
    print(f"{column} is not heavily skewed, skewness value: {skewness_value}")
```
