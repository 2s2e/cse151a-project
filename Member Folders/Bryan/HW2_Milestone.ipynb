Data Encoding

```python
categoricalCols = ['Artist', 'Track', 'Album', 'Album_type', 'Channel', 'Licensed', 'official_video']
```

```python
df['Album_type'].value_counts()
```

```python
# Encode Album_type to be album -> 0, single -> 1, compilation -> 2
def encodeAlbumType(albumType):
  assigned = 0
  if albumType == "album":
    assigned = 0
  elif albumType == "single":
    assigned = 1
  elif albumType == "compilation":
    assigned = 2

  return assigned

df['Album_type'] = df['Album_type'].apply(encodeAlbumType)
```

```python
def encodeBool(value):
  assigned = 0
  if value == True:
    assigned = 1
  elif value == False:
    assigned = 0

  return assigned

# Encode Licensed to be True -> 1, False -> 0
df['Licensed'] = df['Licensed'].apply(encodeBool)
# Encode official_video to be True -> 1, False -> 0
df['official_video'] = df['official_video'].apply(encodeBool)
```
